```python
import matplotlib.pyplot as plt

# 使用plt.imread读取和显示图片的示例
img = plt.imread('demo_image.png')   # img是一个numpy数组，通常shape为(H, W, 3)表示RGB图像，或者(H, W, 4)表示带alpha通道的RGBA图像，类型一般为float32或uint8
plt.imshow(img)           # 显示图像
# plt.axis('off')           # 关闭坐标轴显示
print(f"img的shape是{img.shape}")          # 打印图像的形状，通常为(H, W, 3)或(H, W, 4)
```
